## Arbre de décision classique

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report

# Charger les données
iris = load_iris()
X, y = iris.data, iris.target

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Créer le modèle avec le critère 'gini'
clf_gini = DecisionTreeClassifier(criterion='gini')
clf_gini.fit(X_train, y_train)

# Prédire les étiquettes de test
y_pred_gini = clf_gini.predict(X_test)

# Calculer la précision et afficher le rapport de classification
print(f"Accuracy with Gini: {accuracy_score(y_test, y_pred_gini)}")
print(f"Classification Report with Gini:\n{classification_report(y_test, y_pred_gini)}")


# Charger les données
iris = load_iris()
X, y = iris.data, iris.target

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Créer le modèle avec le critère 'entropy'
clf_entropy = DecisionTreeClassifier(criterion='entropy')
clf_entropy.fit(X_train, y_train)

# Prédire les étiquettes de test
y_pred_entropy = clf_entropy.predict(X_test)

# Calculer la précision et afficher le rapport de classification
print(f"Accuracy with Entropy: {accuracy_score(y_test, y_pred_entropy)}")
print(f"Classification Report with Entropy:\n{classification_report(y_test, y_pred_entropy)}")

# Visualiser les deux arbres de décision en parallèle
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(30, 10))

# Arbre de décision avec le critère 'gini'
plot_tree(clf_gini, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, ax=axes[0])
axes[0].set_title("Decision Tree with Gini Criterion")

# Arbre de décision avec le critère 'entropy'
plot_tree(clf_entropy, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, ax=axes[1])
axes[1].set_title("Decision Tree with Entropy Criterion")

plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report

# Charger les données
iris = load_iris()
X, y = iris.data, iris.target

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Créer le modèle avec le critère 'gini'
clf_gini = DecisionTreeClassifier(criterion='gini')
clf_gini.fit(X_train, y_train)

# Prédire les étiquettes de test
y_pred_gini = clf_gini.predict(X_test)

# Calculer la précision et afficher le rapport de classification
print(f"Accuracy with Gini: {accuracy_score(y_test, y_pred_gini)}")
print(f"Classification Report with Gini:\n{classification_report(y_test, y_pred_gini)}")

# Visualiser l'arbre de décision avec le critère 'gini'
plt.figure(figsize=(20,10))
plot_tree(clf_gini, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree with Gini Criterion")
plt.show()

# Créer le modèle avec le critère 'entropy'
clf_entropy = DecisionTreeClassifier(criterion='entropy')
clf_entropy.fit(X_train, y_train)

# Prédire les étiquettes de test
y_pred_entropy = clf_entropy.predict(X_test)

# Calculer la précision et afficher le rapport de classification
print(f"Accuracy with Entropy: {accuracy_score(y_test, y_pred_entropy)}")
print(f"Classification Report with Entropy:\n{classification_report(y_test, y_pred_entropy)}")

# Visualiser l'arbre de décision avec le critère 'entropy'
plt.figure(figsize=(20,10))
plot_tree(clf_entropy, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree with Entropy Criterion")
plt.show()


# FORCER L ARBRE A ETRE PLUS PETIT et afficher sur la même fenetre 
```

## Forêt aléatoire

```python
from sklearn.ensemble import RandomForestClassifier

# Charger les données
iris = load_iris()
X, y = iris.data, iris.target

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Créer le modèle de forêt aléatoire
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(X_train, y_train)

# Prédire les étiquettes de test
y_pred_rf = rf_clf.predict(X_test)

# Calculer la précision et afficher le rapport de classification
print(f"Accuracy with Random Forest: {accuracy_score(y_test, y_pred_rf)}")
print(f"Classification Report with Random Forest:\n{classification_report(y_test, y_pred_rf)}")

# Visualiser un des arbres de la forêt aléatoire
plt.figure(figsize=(20,10))
plot_tree(rf_clf.estimators_[0], filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Random Forest - One of the Trees")
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Charger les données
iris = load_iris()
X, y = iris.data, iris.target

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Créer le modèle de Gradient Boosting
gb_clf = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_clf.fit(X_train, y_train)

# Prédire les étiquettes de test
y_pred_gb = gb_clf.predict(X_test)

# Calculer la précision et afficher le rapport de classification
print(f"Accuracy with Gradient Boosting: {accuracy_score(y_test, y_pred_gb)}")
print(f"Classification Report with Gradient Boosting:\n{classification_report(y_test, y_pred_gb)}")
# Visualiser un des arbres de Gradient Boosting
plt.figure(figsize=(20,10))
plot_tree(gb_clf.estimators_[0][2], filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Gradient Boosting - One of the Trees")
plt.show()
# Voir comment visualiser les différents arbres
```

## XGBOOST

```python
import xgboost as xgb
from xgboost import XGBClassifier

# Charger les données
iris = load_iris()
X, y = iris.data, iris.target

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Créer le modèle XGBoost
xgb_clf = XGBClassifier(n_estimators=100, eval_metric='logloss')
xgb_clf.fit(X_train, y_train)

# Prédire les étiquettes de test
y_pred_xgb = xgb_clf.predict(X_test)

# Calculer la précision et afficher le rapport de classification
print(f"Accuracy with XGBoost: {accuracy_score(y_test, y_pred_xgb)}")
print(f"Classification Report with XGBoost:\n{classification_report(y_test, y_pred_xgb)}")

```
